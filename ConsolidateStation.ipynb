```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('WeatherNormalized.csv')
df.head(n=20)
```

```python
table = pd.DataFrame()

for i, col in enumerate(df.columns):
    table.insert(i, col, None)
```

```python
regions = [
    ['Ha_Noi', 'Son_Tay', 'Ba_Vi', 'Ha_Dong'],
    ['Hai_Duong', 'Chi_Linh'],
    ['Hung_Yen'],
    ['Nam_Dinh'],
    ['Thai_Binh'],
    ['Ha_Nam', 'Van_Ly'],
    ['Ninh_Binh', 'Nho_Quan']
]

operations = ['avg', 'avg', 'max', 'max', 'avg', 'avg', 'avg', 'max', 'min', 'avg', 'min', 'avg', 'avg', 'max', 'avg', 'max', 'max', 'avg', 'avg']
```

```python
def processing(region, operation, field, year):

    lst = []

    for province in region:

        value = list(df.loc[(df['Year'] == year) & (df['Province'] == province)][field])[0]

        if (not(value == 'x' or pd.isna(value))):
            lst.append(float(value))
    
    
    if (len(lst) == 0):
        return np.nan
    
    lst = np.array(lst)
    
    if (operation == 'avg'):
        return np.mean(lst)
    elif (operation == 'max'):
        return np.max(lst)
    else:
        return np.min(lst)
```

```python
for region in regions:
    for year in range(2000, 2021):
        
        row = [year, region[0]]

        for field_id in range(0, len(df.columns) - 2):
            row.append(processing(region=region, operation=operations[int(field_id/12)], field=df.columns[field_id + 2], year=year))
        
        table.loc[len(table)] = row
        
```

```python
table.to_csv('Weather.csv', index=False, mode='w')
```
