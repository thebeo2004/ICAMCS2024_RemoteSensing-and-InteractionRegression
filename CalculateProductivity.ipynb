Calculating production for farming season

```python
import numpy as np
import pandas as pd
```

```python
area_df = pd.read_csv(r'Data\Winter - Spring\Area.csv')
production_df = pd.read_csv(r'Data\Winter - Spring\Production.csv')
```

```python
provinces = ['Ha_Noi', 'Hai_Duong', 'Hung_Yen', 'Nam_Dinh', 'Ninh_Binh', 'Thai_Binh', 'Ha_Nam']
```

```python
cols = ['Year', 'Province', 'Productivity']

table = pd.DataFrame()

for i, col in enumerate(cols):
    table.insert(i, col, None)
```

```python
table
```

```python
def convert(txt):
    ans, decimal = 0, 0

    for c in txt:
        if (c == ','):
            decimal = 10
            continue
        
        if (decimal == 0):
            ans = ans * 10 + int(c)
        else:
            ans = ans + int(c)/decimal
        
        decimal *= 10
    return ans
```

```python
def get_production(province, year):
    production = convert(list(production_df.loc[production_df['Unnamed: 0'] == province][str(year)])[0])

    if (province == 'Ha_Noi' and year < 2008):
        production += convert(list(production_df.loc[production_df['Unnamed: 0'] == 'Ha_Tay'][str(year)])[0])

    return production
```

```python
def get_area(province, year):
    area = convert(list(area_df.loc[area_df['Unnamed: 0'] == province][str(year)])[0])

    if (province == 'Ha_Noi' and year < 2008):
        area += convert(list(area_df.loc[area_df['Unnamed: 0'] == 'Ha_Tay'][str(year)])[0])

    return area
```

```python
for province in provinces:
    for year in range(2000, 2021):

        row = [year, province]

        production = get_production(province=province, year=year)
        area = get_area(province=province, year=year)

        row.append(production * 10 / area)

        table.loc[len(table)] = row    
```

```python
table.to_csv('CropYield(Winter-Spring).csv', index=False)
```
