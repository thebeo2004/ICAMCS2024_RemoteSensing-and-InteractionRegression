```python
import pandas as pd
```

```python
quantity_df = pd.read_csv(r"Data\RiceQuantity.csv")
area_df = pd.read_csv(r"Data\AreaStatistics.csv")
```

```python
def convert(num):
    ans = 0
    flag = False
    x = 10
    for c in num:
        if (c == '.'):
            continue
        if (c == ','):
            flag = True
            continue
        if (flag):
            ans += int(c)/x
            x *= 10
        else:
            ans = ans * 10 + int(c)
    
    return ans
```

```python
columns = ['Năm', 'TT', 'Diện tích', 'Năng suất']

table = pd.DataFrame()

for i in range(len(columns)):
    table.insert(i, columns[i], None)
```

```python
table
```

```python
provinces = ['Hai_Phong', 'Hung_Yen', 'Nam_Dinh', 'Ninh_Binh', 'Quang_Ninh', 'Thai_Binh', 'Vinh_Phuc', 'Bac_Ninh', 'Ha_Nam', 'Ha_Noi', 'Hai_Duong']
```

```python
for province in provinces:
    query_area = area_df.loc[area_df['TT'] == province]
    query_quantity = quantity_df.loc[quantity_df['TT'] == province]

    for year in range(2000, 2021):

        row = [year, province]

        area = convert(list(query_area[str(year)])[0])
        quantity = convert(list(query_quantity[str(year)])[0])

        if (province == 'Ha_Noi' and year < 2008):
            additional_area = area_df.loc[area_df['TT'] == 'Ha_Tay']
            additional_quantity = quantity_df.loc[quantity_df['TT'] == 'Ha_Tay']
            
            area += convert(list(additional_area[str(year)])[0])
            quantity += convert(list(additional_quantity[str(year)])[0])
        
        row.append(area)
        row.append(quantity)

        table.loc[len(table)] = row
    
```

```python
table.to_csv('Area&Quantity.csv', index=False)
```
